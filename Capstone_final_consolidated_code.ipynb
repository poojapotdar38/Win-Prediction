# Calling packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from termcolor import colored as cl

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Reading File

#win=pd.read_excel(r"D:\data science\Win_Prediction_Data.xlsx")
win=pd.read_excel(r"C:/Users/HP/Desktop/pooja data sci data/project/Win_Prediction_Data.xlsx")
```

# Data cleaning

```python
win.head()
```

```python
win.shape
```

```python
win=win.rename(columns=({"Client Category":"client_category","Solution Type":"solution_type","Deal Date":"deal_date","VP Name":"vp_name","Manager Name":"manager_name","Deal Cost":"deal_cost","Deal Status Code":"deal_status"}))
```

```python
win["combination"]=win["vp_name"]+win["manager_name"]
```

```python
win.head()
```

```python
#Replacing with Mode (Others)
win=win.fillna("others")

#Dropping deal_date column as per analysis in Tableau. (Non-significant)
win=win.drop("deal_date",axis=1)
```

# Data Preparation

```python
deal_log=np.log(win.deal_cost+1)

win=pd.concat([win,deal_log],axis=1)

win.head()
```

```python
win=win.rename(columns=({win.columns[9]:"deal_log"}))

win.head()  #same names for two columns

l1=list(win.columns)

l1[9]="deal_cost_log"

win.columns=l1
```

```python
win.head()
```

```python
win.head()
```

```python
l1=list(win.columns)

l1[8]="combination"

win.columns=l1
```

```python
win.shape
```

```python
vp_manager_info=win
vp_manager_info
```

```python
win=win.drop("vp_name",axis=1)
win=win.drop("manager_name",axis=1)
win=win.drop("deal_log",axis=1)
```

# Converting Non-numeric to Numeric

```python
le=LabelEncoder()
win[win.select_dtypes(include="object").columns]=win[win.select_dtypes(include="object").columns].apply(le.fit_transform)
```

# Creating DF for identifying combinations according to numeric values

```python
vp_manager_info1 = [vp_manager_info["vp_name"], vp_manager_info["manager_name"], win["combination"]]
headers = ["vp_name","manager_name","combination"]
vp_manager_info1 = pd.concat(vp_manager_info1, axis=1, keys=headers)
vp_manager_info2=vp_manager_info1.drop_duplicates()
vp_manager_info_final=vp_manager_info2.sort_values(by=['combination'])
vp_manager_info_final
```

# Random Sampling

```python
win_x=win.drop("deal_status",axis=1)

train_x,test_x,train_y,test_y=train_test_split(win_x,win.deal_status,test_size=0.3,random_state=555)
```

```python
win.deal_cost_log.value_counts()
```

# Model building

```python
#1.Logistic
log_win=LogisticRegression()
log_win.fit(train_x,train_y)
pred_log=log_win.predict(test_x)

#2.Decision tree
dt_win=DecisionTreeClassifier()
dt_win.fit(train_x,train_y)
pred_dt=dt_win.predict(test_x)

#3.Random forest
rf_win=RandomForestClassifier()
rf_win.fit(train_x,train_y)
pred_rf=rf_win.predict(test_x)

#4.K-Neareast neighbour
knn_win=KNeighborsClassifier(n_neighbors=5)
knn_win.fit(train_x,train_y)
pred_knn=knn_win.predict(test_x)
```

# Evaluation

```python
# Evaluation

from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

print(cl("Confusion matrix for Logistic model is",attrs=['bold'],color='yellow'))
print(cl("{}".format(confusion_matrix(pred_log,test_y)),attrs=['bold'],color='yellow'))
print(cl("Accuracy for Logistic model is        {}".format(accuracy_score(pred_log,test_y)),attrs=['bold'],color='yellow'))
print(cl("Precision for Logistic model is       {}".format(precision_score(pred_log,test_y)),attrs=['bold'],color='yellow'))
print(cl("F1 Score for Logistic model is        {}".format(f1_score(pred_log,test_y)),attrs=['bold'],color='yellow'))
print(cl("AU-ROC Score for Logistic model is    {}".format(roc_auc_score(pred_log,test_y)),attrs=['bold'],color='yellow'))

print(cl("==============================================================",attrs=['bold']))
print(cl("Confusion matrix for Decision tree model is",attrs=['bold']))
print(cl("{}".format(confusion_matrix(pred_dt,test_y)),attrs=['bold']))
print(cl("Accuracy for Decision tree model is        {}".format(accuracy_score(pred_dt,test_y)),attrs=['bold']))
print(cl("Precision for Decision tree model is       {}".format(precision_score(pred_dt,test_y)),attrs=['bold']))
print(cl("F1 Score for Decision tree model is        {}".format(f1_score(pred_dt,test_y)),attrs=['bold']))
print(cl("AU-ROC Score for Decision tree model is    {}".format(roc_auc_score(pred_dt,test_y)),attrs=['bold']))

print(cl("==============================================================",attrs=['bold']))
print(cl("Confusion matrix for Random forest model is",attrs=['bold'],color='green'))
print(cl("{}".format(confusion_matrix(pred_rf,test_y)),attrs=['bold'],color='green'))
print(cl("Accuracy for Random forest model is        {}".format(accuracy_score(pred_rf,test_y)),attrs=['bold'],color='green'))
print(cl("Precision for Random forest model is       {}".format(precision_score(pred_rf,test_y)),attrs=['bold'],color='green'))
print(cl("F1 Score for Random forest model is        {}".format(f1_score(pred_rf,test_y)),attrs=['bold'],color='green'))
print(cl("AU-ROC Score for Random forest model is    {}".format(roc_auc_score(pred_rf,test_y)),attrs=['bold'],color='green'))

print(cl("==============================================================",attrs=['bold']))
print(cl("Confusion matrix for K-NN model is",attrs=['bold'],color='blue'))
print(cl("{}".format(confusion_matrix(pred_knn,test_y)),attrs=['bold'],color='blue'))
print(cl("Accuracy for K-NN model is        {}".format(accuracy_score(pred_knn,test_y)),attrs=['bold'],color='blue'))
print(cl("Precision for K-NN model is       {}".format(precision_score(pred_knn,test_y)),attrs=['bold'],color='blue'))
print(cl("F1 Score for K-NN model is        {}".format(f1_score(pred_knn,test_y)),attrs=['bold'],color='blue'))
print(cl("AU-ROC Score for K-NN model is    {}".format(roc_auc_score(pred_knn,test_y)),attrs=['bold'],color='blue'))
```

# For finding top 5 combination

Creating data frames containing probability

```python
pred_knn_proba=knn_win.predict_proba(test_x)
pred_knn_proba=pd.DataFrame(pred_knn_proba)

pred_log_proba=log_win.predict_proba(test_x)
pred_log_proba=pd.DataFrame(pred_log_proba)

pred_dt_proba=dt_win.predict_proba(test_x)
pred_dt_proba=pd.DataFrame(pred_dt_proba)

pred_rf_proba=rf_win.predict_proba(test_x)
pred_rf_proba=pd.DataFrame(pred_rf_proba)
```

Creating data frames containing win probability and loss probability

```python
test_data = pd.DataFrame(test_x)
import math
test_data['deal_cost']=test_data.apply(lambda row:np.exp(row.deal_cost_log)-1,axis=1)
test_data=pd.concat([test_data, test_y.rename('deal_status_original')], axis=1)

pred_knn_df=pd.DataFrame(pred_knn)
test_data_knn= pd.concat([test_data.reset_index(drop=True), pred_knn_df], axis=1)
test_data_knn.columns.values[-1]="as_predicted"
test_data_knn = pd.concat([test_data_knn.reset_index(drop=True), pred_knn_proba], axis=1)
test_data_knn.columns.values[-1]="win_prob"
test_data_knn.columns.values[-2]="loss_prob"


pred_log_df=pd.DataFrame(pred_log)
test_data_log= pd.concat([test_data.reset_index(drop=True), pred_log_df], axis=1)
test_data_log.columns.values[-1]="as_predicted"
test_data_log = pd.concat([test_data_log.reset_index(drop=True), pred_log_proba], axis=1)
test_data_log.columns.values[-1]="win_prob"
test_data_log.columns.values[-2]="loss_prob"


pred_dt_df=pd.DataFrame(pred_dt)
test_data_dt= pd.concat([test_data.reset_index(drop=True), pred_dt_df], axis=1)
test_data_dt.columns.values[-1]="as_predicted"
test_data_dt = pd.concat([test_data_dt.reset_index(drop=True), pred_dt_proba], axis=1)
test_data_dt.columns.values[-1]="win_prob"
test_data_dt.columns.values[-2]="loss_prob"


pred_rf_df=pd.DataFrame(pred_rf)
test_data_rf= pd.concat([test_data.reset_index(drop=True), pred_rf_df], axis=1)
test_data_rf.columns.values[-1]="as_predicted"
test_data_rf = pd.concat([test_data_rf.reset_index(drop=True), pred_rf_proba], axis=1)
test_data_rf.columns.values[-1]="win_prob"
test_data_rf.columns.values[-2]="loss_prob"
```

Finding weighted average of Win probability.

```python
print(cl("=============================================================="))
test_data_log['deal_costXwin_prob']=test_data_log.apply(lambda row:row.win_prob*row.deal_cost,axis=1)
test_data_sum_log = test_data_log.groupby('combination', as_index=False).sum()
test_data_sum_log['weighted_average']=test_data_sum_log.apply(lambda row:row.deal_costXwin_prob/row.deal_cost,axis=1)
test_data_sum_wtavg_log = test_data_sum_log.merge(vp_manager_info_final, on='combination', how='inner')
test_data_sum_wtavg_log["weighted_average"] = pd.to_numeric(test_data_sum_wtavg_log.weighted_average, errors='coerce')
test_data_sum_wtavg_log_sort=test_data_sum_wtavg_log.sort_values(by='weighted_average', ascending=False)
print(cl("Top five combination of VP & Manager for Logistic model",attrs=['bold'],color='yellow'))
print(cl(test_data_sum_wtavg_log_sort[['vp_name','manager_name']].head(5),attrs=['bold'],color='yellow'))

print(cl("=============================================================="))

test_data_dt['deal_costXwin_prob']=test_data_dt.apply(lambda row:row.win_prob*row.deal_cost,axis=1)
test_data_sum_dt = test_data_dt.groupby('combination', as_index=False).sum()
test_data_sum_dt['weighted_average']=test_data_sum_dt.apply(lambda row:row.deal_costXwin_prob/row.deal_cost,axis=1)
test_data_sum_wtavg_dt = test_data_sum_dt.merge(vp_manager_info_final, on='combination', how='inner')
test_data_sum_wtavg_dt["weighted_average"] = pd.to_numeric(test_data_sum_wtavg_dt.weighted_average, errors='coerce')
test_data_sum_wtavg_dt_sort=test_data_sum_wtavg_dt.sort_values(by='weighted_average', ascending=False)
print(cl("Top five combination of VP & Manager for Decision Tree model",attrs=['bold']))
print(cl(test_data_sum_wtavg_dt_sort[['vp_name','manager_name']].head(5),attrs=['bold']))

print(cl("=============================================================="))

test_data_rf['deal_costXwin_prob']=test_data_rf.apply(lambda row:row.win_prob*row.deal_cost,axis=1)
test_data_sum_rf = test_data_rf.groupby('combination', as_index=False).sum()
test_data_sum_rf['weighted_average']=test_data_sum_rf.apply(lambda row:row.deal_costXwin_prob/row.deal_cost,axis=1)
test_data_sum_wtavg_rf = test_data_sum_rf.merge(vp_manager_info_final, on='combination', how='inner')
test_data_sum_wtavg_rf["weighted_average"] = pd.to_numeric(test_data_sum_wtavg_rf.weighted_average, errors='coerce')
test_data_sum_wtavg_rf_sort=test_data_sum_wtavg_rf.sort_values(by='weighted_average', ascending=False)
print(cl("Top five combination of VP & Manager for Random Forest model",attrs=['bold'],color='green'))
print(cl(test_data_sum_wtavg_rf_sort[['vp_name','manager_name']].head(5),attrs=['bold'],color='green'))

print(cl("=============================================================="))

test_data_knn['deal_costXwin_prob']=test_data_knn.apply(lambda row:row.win_prob*row.deal_cost,axis=1)
test_data_sum_knn = test_data_knn.groupby('combination', as_index=False).sum()
test_data_sum_knn['weighted_average']=test_data_sum_knn.apply(lambda row:row.deal_costXwin_prob/row.deal_cost,axis=1)
test_data_sum_wtavg_knn = test_data_sum_knn.merge(vp_manager_info_final, on='combination', how='inner')
test_data_sum_wtavg_knn["weighted_average"] = pd.to_numeric(test_data_sum_wtavg_knn.weighted_average, errors='coerce')
test_data_sum_wtavg_knn_sort=test_data_sum_wtavg_knn.sort_values(by='weighted_average', ascending=False)
print(cl("Top five combination of VP & Manager for K-Neareast neighbour model",attrs=['bold'],color='blue'))
print(cl(test_data_sum_wtavg_knn_sort[['vp_name','manager_name']].head(5),attrs=['bold'],color='blue'))
print(cl("=============================================================="))
```

```python
test_data_log['false_pred']=test_data_log.apply(lambda row:"False" if row.deal_status_original == row.as_predicted else "True",axis=1)
cost_of_false_pred=test_data_log[test_data_log["false_pred"]== "True"].sum()["deal_cost"]/test_data_log[test_data_log["false_pred"]== "True"].count()["deal_cost"]
print(cl("Cost of False Prediction for Logistic model is           {}".format(np.round(cost_of_false_pred,2)),attrs=['bold'],color='yellow'))
print(cl("========================================================================"))

test_data_dt['false_pred']=test_data_dt.apply(lambda row:"False" if row.deal_status_original == row.as_predicted else "True",axis=1)
cost_of_false_pred=test_data_dt[test_data_dt["false_pred"]== "True"].sum()["deal_cost"]/test_data_dt[test_data_dt["false_pred"]== "True"].count()["deal_cost"]
print(cl("Cost of False Prediction for Decision tree model is      {}".format(np.round(cost_of_false_pred,2)),attrs=['bold']))
print(cl("========================================================================"))


test_data_dt['false_pred']=test_data_dt.apply(lambda row:"False" if row.deal_status_original == row.as_predicted else "True",axis=1)
cost_of_false_pred=test_data_dt[test_data_dt["false_pred"]== "True"].sum()["deal_cost"]/test_data_dt[test_data_dt["false_pred"]== "True"].count()["deal_cost"]
print(cl("Cost of False Prediction for Random forest is            {}".format(np.round(cost_of_false_pred,2)),attrs=['bold'],color='green'))
print(cl("========================================================================"))


test_data_knn['false_pred']=test_data_knn.apply(lambda row:"False" if row.deal_status_original == row.as_predicted else "True",axis=1)
cost_of_false_pred=test_data_knn[test_data_knn["false_pred"]== "True"].sum()["deal_cost"]/test_data_knn[test_data_knn["false_pred"]== "True"].count()["deal_cost"]
print(cl("Cost of False Prediction for KNN model is                {}".format(np.round(cost_of_false_pred,2)),attrs=['bold'],color='blue'))
print(cl("========================================================================"))
```
